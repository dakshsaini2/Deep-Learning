```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("placement (2).csv")
```

```python
df.shape
df.head()
```

```python

plt.figure(figsize=(8,6))

sns.scatterplot(
    x=df['cgpa'],
    y=df['resume_score'],
    hue=df['placed'],
    palette='coolwarm'
)

plt.title("CGPA vs Resume Score")

plt.show()
```

```python
X=df.iloc[:,0:2]
y=df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
```

```python
p.fit(X,y)
```

```python
p.coef_
```

```python
p.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X.values,y.values,clf=p,legend=2)
```
